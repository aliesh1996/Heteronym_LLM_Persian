```python
newsentence = 'خرد و بینش دانشمندان ره سر منزل مقصود به ما آموزد'



new_words = newsentence.split()

for word in new_words:

  if word in heteronyms:

    pronunciations = []

    texts = []





    texts.append([newsentence])



    for sentence in heter_dict[word]:



      s = sentence.split()
################################
      sub_sentences = []
      section = []


      r_sentence = s[1:len(s)-2]

      pronunciations.append(s[len(s)-1])

      # print(sentence)

      # print(s[len(s)-1])



      for c in r_sentence:

        if c != '-':
          section.append(c)
        else:
          sub_sentences.append(section)
          section = []


      # print(sub_sentences)

      the_sentences = []

      for lst in sub_sentences:





        s_sentence = ''

        for i in lst:

          if i != lst[len(lst)-1]:
            s_sentence = s_sentence + i + ' '

          else:
            s_sentence = s_sentence + i

        the_sentences.append(s_sentence)


      texts.append(the_sentences)

      print(texts)
      print(pronunciations)

# print(texts)

    allsegmenttensors = []
    alltokens = []


    for text in texts:


      if len(text) == 1:

        # print(text)



        segments_ids = [1] * len(tokenizer.tokenize(text[0]))

        tokens_tensor = torch.tensor([tokenizer.convert_tokens_to_ids(tokenizer.tokenize(text[0]))])

        segments_tensors = torch.tensor([segments_ids])

        alltokens.append(tokens_tensor)

        allsegmenttensors.append(segments_tensors)

      else:

        segmenttensors = []
        tokens = []

        for i in text:

          segments_ids = [1] * len(tokenizer.tokenize(i))

          tokens_tensor = torch.tensor([tokenizer.convert_tokens_to_ids(tokenizer.tokenize(i))])

          segments_tensors = torch.tensor([segments_ids])

          tokens.append(tokens_tensor)

          segmenttensors.append(segments_tensors)

        alltokens.append(tokens)

        allsegmenttensors.append(segmenttensors)







    hiddens = []

    for i in range(len(alltokens)):

      with torch.no_grad():

        if len(alltokens[i]) == 1:

          outputs = model(alltokens[i], allsegmenttensors[i])

          # Evaluating the model will return a different number of objects based on
          # how it's  configured in the `from_pretrained` call earlier. In this case,
          # becase we set `output_hidden_states = True`, the third item will be the
          # hidden states from all layers. See the documentation for more details:
          # https://huggingface.co/transformers/model_doc/bert.html#bertmodel
          hidden_states = outputs[2]

          hiddens.append([hidden_states])

        else:

          hiddens_in = []

          for j in range(len(alltokens[i])):

            # print(alltokens[i][j].long())

            outputs = model(alltokens[i][j].long(), allsegmenttensors[i][j].long())

          # Evaluating the model will return a different number of objects based on
          # how it's  configured in the `from_pretrained` call earlier. In this case,
          # becase we set `output_hidden_states = True`, the third item will be the
          # hidden states from all layers. See the documentation for more details:
          # https://huggingface.co/transformers/model_doc/bert.html#bertmodel
            hidden_states = outputs[2]

            hiddens_in.append(hidden_states)

          hiddens.append(hiddens_in)

    # print(len(hiddens))


    # token_embeddings = []

    token_embeddings = []

    sentence_vecs = []


    summations = []

    # sentence_embeddings = []

    for hidden_states in hiddens:

      if len(hidden_states) == 1:


        token_embeddings = torch.stack(hidden_states[0], dim=0)

        token_vecs = hidden_states[0][-2][0]

        sentence_embedding = torch.mean(token_vecs, dim=0)

        #token_embeddings.size()

        token_embeddings = torch.squeeze(token_embeddings, dim=1)

        #token_embeddings.size()

        token_embeddings = token_embeddings.permute(1,0,2)

        token_vecs_sum = []

        sentence_vecs.append([sentence_embedding])







        for token in token_embeddings:

          # `token` is a [12 x 768] tensor

          # Concatenate the vectors (that is, append them together) from the last
          # four layers.
          # Each layer vector is 768 values, so `cat_vec` is length 3,072.
          # cat_vec = torch.cat((token[-1], token[-2], token[-3], token[-4]), dim=0)
          sum_vec = torch.sum(token[-4:], dim=0)
          token_vecs_sum.append(sum_vec)



        summations.append([token_vecs_sum[-1]])

      else:

        token_embeddings_in = []


        summations_in = []

        sentence_vecs_in = []

        for h in hidden_states:

          token_embeddings = torch.stack(h, dim=0)

          token_vecs = h[-2][0]

          sentence_embedding = torch.mean(token_vecs, dim=0)

          #token_embeddings.size()

          token_embeddings = torch.squeeze(token_embeddings, dim=1)

          #token_embeddings.size()

          token_embeddings = token_embeddings.permute(1,0,2)

          token_vecs_sum = []





          for token in token_embeddings:

            # `token` is a [12 x 768] tensor

            # Concatenate the vectors (that is, append them together) from the last
            # four layers.
            # Each layer vector is 768 values, so `cat_vec` is length 3,072.
            # cat_vec = torch.cat((token[-1], token[-2], token[-3], token[-4]), dim=0)
            sum_vec = torch.sum(token[-4:], dim=0)
            token_vecs_sum.append(sum_vec)


          summations_in.append(token_vecs_sum[-1])

          sentence_vecs_in.append(sentence_embedding)

        summations.append(summations_in)
        sentence_vecs.append(sentence_vecs_in)

#####################################################
    # print(summations[0])

    results = []

    for i in range(1,len(sentence_vecs)):

      mean = 0

      for j in sentence_vecs[i]:

        mean += 1 - spatial.distance.cosine(np.array(sentence_vecs[0][0]), np.array(j))

      mean = mean/len(sentence_vecs[i])

      results.append(mean)



      # print(results)

    results = np.array(results)

    # print(results)

    maxElement = np.amax(results)

    maxindex = np.where(results == np.amax(results))

    # print(maxindex)

    # print(pronunciations[int(maxindex[0][0])])




    for n, i in enumerate(new_words):
      if i == word:
       new_words[n] = pronunciations[int(maxindex[0][0])]
      #  print(pronunciations[-1])

# print(new_words)

final_sentence = ''

for i in new_words:

    if i != len(new_words)-1:
      final_sentence = final_sentence + i + ' '

    else:
      final_sentence = final_sentence + i

print(final_sentence)
# print(texts)
# print(pronunciations)
```
